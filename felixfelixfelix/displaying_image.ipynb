```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# --- Configuration: Update these paths ---

# 1. Set the root path to your dataset (same as your YOLO_DATASET_ROOT_PATH)
#    Ensure this path points to where your dataset is on Vertex AI Workbench
#    (e.g., if it's in Google Drive and mounted at /content/drive/MyDrive/)
dataset_root_path = "/home/jupyter/felixfelixfelix/yolo_dataset_output" # Or "/content/drive/MyDrive/your_dataset_folder"

# 2. Specify the subfolder (e.g., 'train/images' or 'val/images')
image_subfolder = "train/images" # Or "val/images", or "test/images"

# 3. Specify the exact filename of the JPG image you want to open
image_filename = "15534.jpg" # REPLACE THIS with an actual JPG filename from your dataset

# --- Code to open and display the image ---

# Construct the full path to the image
full_image_path = os.path.join(dataset_root_path, image_subfolder, image_filename)

print(f"Attempting to open image: {full_image_path}")

try:
    # Open the image using Pillow (PIL)
    img = Image.open(full_image_path)

    # Display the image using Matplotlib
    plt.figure(figsize=(10, 8)) # Adjust figure size as needed
    plt.imshow(img)
    plt.title(f"Displaying: {image_filename}")
    plt.axis('off')  # Turn off axis numbers and ticks
    plt.show()

    print(f"\nImage details:")
    print(f"  Format: {img.format}")
    print(f"  Size (Width x Height): {img.size}")
    print(f"  Mode: {img.mode}")

except FileNotFoundError:
    print(f"ERROR: Image not found at '{full_image_path}'")
    print("Please check the following:")
    print(f"  1. Is '{dataset_root_path}' the correct root path to your dataset?")
    print(f"  2. Is the image located in the '{image_subfolder}' subfolder?")
    print(f"  3. Is '{image_filename}' the correct filename (including .jpg extension and case sensitivity)?")
except Exception as e:
    print(f"An error occurred while trying to open or display the image: {e}")
```
